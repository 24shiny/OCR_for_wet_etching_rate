```python
# interpolation w.r.t both directions
new_col = np.arange(w.columns[0], w.columns[-1]+0.25, 0.25)
w_inter = pd.DataFrame(columns=new_col, index=w.index)
for idx, row in w.iterrows():
    w_inter.loc[idx] = np.interp(new_col, w.columns.values, w.loc[idx])
w_inter
```
