```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

e1 = pd.read_csv('1_E.csv', index_col='cm')
e1 = e1.iloc[1:]
e1.index = [5+i for i in range(len(e1))]
e1 = e1 / 1000

w1 = pd.read_csv('1_W.csv', index_col='cm')
w1 = w1.iloc[1:]
w1.index = [5+i for i in range(len(w1))]
w1 = w1 / 1000
```

```python
w2 = pd.read_csv('2_W.csv', index_col='cm')
e2 = pd.read_csv('2_E.csv', index_col='cm')
t2 = pd.read_csv('2_T.csv', index_col='cm')
b2 = pd.read_csv('2_B.csv', index_col='cm')
```

```python
# drop the first column
w2 = w2.iloc[1:]
e2 = e2.iloc[1:]
t2 = t2.iloc[1:]
b2 = b2.iloc[1:]

new_index = [4,5,5.25,5.5,5.75,6,7,8,9,10,11]
w2.index = new_index
e2.index = new_index
t2.index = new_index
b2.index = new_index

w2 = w2 / 1000
e2 = e2 / 1000
t2 = t2 / 1000
b2 = b2 /1000
```

```python
t3 = pd.read_csv('3_T.csv', index_col='cm')
b3 = pd.read_csv('3_B.csv', index_col='cm')
```

```python
t3 = t3.iloc[1:]

new_index = [4,5,5.25,5.5,5.75,6,7,8,9,10,11]
t3.index = new_index
b3.index = new_index

t3 = t3 / 1000
b3 = b3 / 1000
```

Customize their layout

```python
fig, axs = plt.subplots()
sns.heatmap(e1, annot=True, cmap=plt.cm.GnBu, fmt=".1f", cbar_kws={'label':'Thickness [um]'}, vmax=760, vmin=740)
plt.title('E_1')
axs.set(xlabel='X axis [cm]', ylabel='Y axis [cm]')
plt.savefig('1_E_plot')
plt.show()

fig, axs = plt.subplots()
sns.heatmap(e1-e1.min().min(), annot=True, cmap=plt.cm.GnBu, fmt=".1f", cbar_kws={'label':'Relative thickness [um]'}, vmax=20, vmin=0)
plt.title('E_1')
axs.set(xlabel='X axis [cm]', ylabel='Y axis [cm]')
plt.savefig('1_E_relative_plot')
plt.show()
```

```python
df_to_plot = w1
name = '1_W'
fig, axs = plt.subplots()
sns.heatmap(df_to_plot, annot=True, cmap=plt.cm.GnBu, fmt=".0f", cbar_kws={'label':'Thickness [um]'}, vmax=760, vmin=740)
plt.title(f'{name}')
axs.set(xlabel='X axis [cm]', ylabel='Y axis [cm]')
plt.savefig(f'{name}_plot')
plt.show()

fig, axs = plt.subplots()
sns.heatmap(df_to_plot-df_to_plot.min().min(), annot=True, cmap=plt.cm.GnBu, fmt=".0f", cbar_kws={'label':'Relative thickness [um]'}, vmax=20, vmin=0)
plt.title(f'{name}')
axs.set(xlabel='X axis [cm]', ylabel='Y axis [cm]')
plt.savefig(f'{name}_relative_plot')
plt.show()
```
